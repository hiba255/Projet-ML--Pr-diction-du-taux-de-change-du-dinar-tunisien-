### Dataset: Balance Commerciale

```python
import pandas as pd
balance_commerciale_df = pd.read_excel('/content/datasets2/final_balance_commerciale_data.xlsx')
display(balance_commerciale_df.head())
```

### Filtering `balance_commerciale_df`

```python
balance_commerciale_df['date'] = pd.to_datetime(balance_commerciale_df['date'])
balance_commerciale_df_filtered = balance_commerciale_df[(balance_commerciale_df['date'] >= '2015-01-01') & (balance_commerciale_df['date'] <= '2026-01-01')]
display(balance_commerciale_df_filtered.head())
display(balance_commerciale_df_filtered.tail())
```

```python
balance_commerciale_df_filtered.set_index('date', inplace=True)
```

```python
df_daily = balance_commerciale_df_filtered.resample('D').interpolate(method='linear')
```

```python
display(df_daily.head())
display(df_daily.tail())
```

```python
df_daily = df_daily.reset_index()
```

```python
df_daily.to_excel('balance_commerciale_daily.xlsx', index=False)
```
