```python
import pandas as pd

# Lire le fichier
taux_interet_df = pd.read_excel('/content/datasets2/taux_interet.xlsx')

# Supprimer les espaces autour des noms de colonnes
taux_interet_df.columns = taux_interet_df.columns.str.strip()

# Renommer la colonne année pour plus de clarté
taux_interet_df.rename(columns={'Date/Indicateurs':'year'}, inplace=True)

# Filtrer juste de 2015 à 2025 (inclus)
taux_interet_df = taux_interet_df[(taux_interet_df['year'].dt.year >= 2015) &
                                  (taux_interet_df['year'].dt.year <= 2025)]

taux_interet_df.head()
```

```python
# Transformer “large → long”
df_long = taux_interet_df.melt(
    id_vars=['year'],
    value_vars=['Janvier','Février','Mars','Avril','Mai','Juin',
                'Juillet','Août','Septembre','Octobre','Novembre','Décembre'],
    var_name='month',
    value_name='taux_interet'
)
```

```python
# Dictionnaire mois français → numéro
mois_dict = {
    'Janvier':1, 'Février':2, 'Mars':3, 'Avril':4, 'Mai':5, 'Juin':6,
    'Juillet':7, 'Août':8, 'Septembre':9, 'Octobre':10, 'Novembre':11, 'Décembre':12
}

df_long['month_num'] = df_long['month'].map(mois_dict)

# Créer une date au premier jour du mois
df_long['date'] = pd.to_datetime(dict(
    year=df_long['year'].dt.year,
    month=df_long['month_num'],
    day=1
))

# Garder uniquement les colonnes utiles
df_long = df_long[['date','taux_interet']].sort_values('date').reset_index(drop=True)
df_long.head()
```

```python
df_long
```

```python
# Mettre date en index pour resample
df_daily = df_long.set_index('date').resample('D').ffill().reset_index()
df_daily.head(10)
```

```python
df_daily
```

```python
# Filtrer exactement entre 2015-01-01 et 2026-01-01
df_daily = df_daily[(df_daily['date'] >= '2015-01-01') & (df_daily['date'] <= '2026-01-01')]
df_daily.head()
df_daily.tail()
```

```python
df_daily.to_excel('taux_interet_daily.xlsx', index=False)
```
